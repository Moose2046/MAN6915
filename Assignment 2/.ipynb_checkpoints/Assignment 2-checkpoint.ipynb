# Problem 1 (5 points)

Write a simple program to compute the tax on an order amount. The program should prompt for the order amount and the state. If the state is “WI,” then the order must be charged 5.5% tax. The program should display the subtotal, tax, and total for Wisconsin residents but display just the total for non-residents.

## Example Output
<pre>
What is the order amount? 10
What is the state? WI
The subtotal is $10.00.
The tax is $0.55.
The total is $10.55.
</pre>

Or

<pre>
What is the order amount? 10
What is the state? MN
The total is $10.00
</pre>

## Constraints

- Implement this program using only a simple if statement—don’t use an else clause.
- Ensure that all money is rounded up to the nearest cent.
- Use a single output statement at the end of the program to display the program results.

```python
from decimal import Decimal, ROUND_CEILING

def problem1():
    #WI Tax Rate
    WI_TAX_RATE=0.055
    
    #Input
    orderAmountString = input("What is the order amount? ")
    stateString = input("What is the state? ")
    
    #Conversions 
    orderAmountFloat = float(orderAmountString)
    totalFloat = orderAmountFloat
    outputString = ""

    #Checks to see if state is WI if so then gets taxes and total 
    if stateString == "WI":
        
        taxFloat = WI_TAX_RATE * orderAmountFloat
        totalFloat = totalFloat + taxFloat

        #Creates output strings for additional calculations 
        subTotalString = "The subtotal is ${:.2f}".format(orderAmountFloat)
        taxString = "The tax is ${:.2f}".format(taxFloat)
        outputString = "".join([subTotalString,'\n', taxString, '\n'])
    
    #Gets the total Cost and creates an output string
    totalRounded = Decimal(totalFloat).quantize(Decimal('.001'), rounding=ROUND_CEILING)
    totalString = "The total is ${:.2f}".format(totalRounded)
    
    outputString = ''.join([outputString, totalString])

    print(outputString)    

problem1()
```

### Problem 4 (3 points)

Create a program that prompts for your weight, gender, number of drinks, the amount of alcohol by volume of the drinks consumed, and the amount of time since your last drink. Calculate your blood alcohol content (BAC) using this formula $BAC = (A \times 5.14 / W \times r) - 0.015 \times H$

where

- A is total alcohol consumed, in ounces (oz).
- W is body weight in pounds.
- r is the alcohol distribution ratio:
  - 0.73 for men
  - 0.66 for women
- H is number of hours since the last drink.

Display whether or not it’s legal to drive by comparing the blood alcohol content to 0.08.

## Example Output
<pre>
Your BAC is 0.08
It is not legal for you to drive.
</pre>

## Constraint

- Prevent the user from entering non-numeric values.

```python
#Define Constants
ALCOHOL_RATIO_MEN = 0.73
ALCOHOL_RATIO_WOMEN = 0.66

def safeToDrive():

    #Gets weight input
    while True:
        try:
            weightString = input("Input your weight: ")
            weightFloat = float(weightString)
            break;
        except:
            print("Input must be a number")
    #Gets Gender
    while True:
        genderString = input("Input your gender(M/F): ")
        if(genderString == 'M'):
            alcoholRatio = ALCOHOL_RATIO_MEN
            break;
        elif(genderString =='F'):
            alcoholRatio = ALCOHOL_RATIO_WOMEN
            break;
        else:
            print("Input a valid gender") 
    #Gets number of drinks
    while True:
        try:
            drinkNumString = input("Input number of drinks: ")
            drinkNumFloat = float(drinkNumString)
            break;
        except:
            print("Input must be a number")        
    #Gets Number of drinks 
    while True:
        try:
            drinkVolumeString = input("Input the size of drinks in ounces: ")
            drinkVolumeFloat = float(drinkVolumeString)
            break;
        except:
            print("Input must be a number")        
    #Gets how long since the last drink
    while True:
        try:
            timeSinceLastDrinkString = input("Input how many hours since your last drink: ")
            timeSinceLastDrinkInHoursFloat = float(timeSinceLastDrinkString)
            break;
        except:
            print("Input must be a number")
    #Total ounces consumed
    totalConsumpution = drinkVolumeFloat * drinkNumFloat

    #Calculates BAC Value
    bacFloat = (totalConsumpution * 5.14 / weightFloat * alcoholRatio ) - (0.015 * timeSinceLastDrinkInHoursFloat )
    if (bacFloat < 0):
        bacFloat = 0
    #Output
    outputString = "Your BAC is {0:.2f}\n".format(bacFloat) 
    if(bacFloat >= 0.08):
        outputString +="It is not legal for you to drive"
    else:
        outputString += "It is legal for you to drive"

    print(outputString)
safeToDrive()    

```

```python
def tempConvert():
    #Keeps asking for valid C or F until one is input
    while True:
            print("Press C to convert from Fahrenheit to Celsius.\nPress F to convert from Celsius to Fahrenheit.")
            
            choiceString = input("Your choice: ")
            choiceUpperString = choiceString.upper()
            
            #Converts from C to F
            if(choiceUpperString =='C'):
                temperature = getTemp('Fahrenheit')
                convertedTempFloat = (temperature -32) * (5/9)
                printTemp(convertedTempFloat, 'Celsius')
                break
            #converts from F to C 
            elif(choiceUpperString == 'F'):
                temperature = getTemp('Celsius')
                convertedTempFloat = (temperature *(9/5)) + 32
                printTemp(convertedTempFloat, 'Fahrenheit')
                break
            else:
                print("Input a valid choice!")
#Gets the temp and returns it
#Used to print proper message and reduce number of input statements needed. 
def getTemp(tempType):
    while True:
        try:
            tempString = input("".join(["Please enter the temperature in ", tempType, " "]))
            tempFloat = float(tempString)
            break
        except ValueError as e:
            print("Please input a valid number!")
    return tempFloat
#prints final output
def printTemp(temp, tempType):
    print("The temperature in {0} is {1:.0f}".format(tempType, temp))
tempConvert()
```

Create a program to calculate the body mass index (BMI) for a person using the person’s height in inches and weight in pounds. The program should prompt the user for weight and height.

Calculate the BMI by using the following formula: $bmi = (weight / (height \times height)) \times 703$

If the BMI is between 18.5 and 25, display that the person is at a normal weight. If they are out of that range, tell them if they are underweight or overweight and tell them to consult their doctor.

## Example Output

<pre>

or

<pre>
Your BMI is 32.5.
You are overweight. You should see your doctor.

- Ensure your program takes only numeric data. Don’t let the user continue unless the data is valid.

```python
def bmiCheck():
    #Gets weight in pounds
    while True:
        try:
            weightString = input("Input your weight in pounds: ")
            weightFloat = float(weightString)
            break;
        except:
            print("Input must be a number")
    #Gets height in inches
    while True:
        try:
            heightString = input("Input your height in inches: ")
            heightFloat = float(heightString)
            break;
        except:
            print("Input must be a number")
    #Calcs bmi
    bmi = (weightFloat / (heightFloat*heightFloat)) * 703
    
    #output section
    output = "Your BMI is {0:.1f}.\n".format(bmi)
    if(bmi <18.5):
        output += "You are underweight. You should see your doctor"
    elif(bmi<25):
        output += "You are within the ideal weight range"
    else:
        output += "You are overweight. You should see your doctor"
    print (output)

bmiCheck()
```

Create a tax calculator that handles multiple states and multiple counties within each state. The program prompts the user for the order amount and the state where the order will be shipped.

Wisconsin residents are charged a base tax of 5%. Wisconsin residents are also charged additional tax based on their county of residence.
- For Eau Claire county residents, add an additional 0.005 tax.
- For Dunn county residents, add an additional 0.004 tax.

Illinois residents must be charged 8% sales tax with no additional county-level charge. All other states are not charged tax. The program then displays the tax and the total for Wisconsin and Illinois residents but just the total for everyone else.

## Example Output

<pre>
What is the order amount? 10
What state do you live in? Wisconsin
What county do you live in? Eau Claire
The tax is $0.55.
The total is $10.55.
</pre>

## Constraints

- Ensure that all money is rounded up to the nearest cent.
- Use a single output statement at the end of the program to display the program results.

```python
#Imports
from decimal import Decimal, ROUND_CEILING

def calcTaxes():
    #WI Tax Rate
    WI_TAX_RATE=0.05
    EAU_TAX_RATE=0.005
    DUNN_TAX_RATE=0.004
    IL_TAX_RATE = 0.08
    additionTaxOutput = False
    #Input
    #Verifies orderAmount is a number
    while True:
        try:
            orderAmountString = input("What is the order amount? ")
            orderAmountFloat = float(orderAmountString)
            break;
        except:
            print("Input must be a number")
    
    #Gets State and country
    stateString = input("What state do you live in?  ")
    countyString = input("What county do you live in? ")
    
    #Just assigns order amount to total incase no additional calculations are needed
    totalFloat = orderAmountFloat
    outputString = ""

    #Checks to see if we need to calculate taxes for Illinois or Wisconsin
    if stateString == 'Illinois':
        additionTaxOutput = True
        taxRateFloat = IL_TAX_RATE
    elif stateString == "Wisconsin" :
        additionTaxOutput = True
        taxRateFloat = WI_TAX_RATE
        #Checks the counties for additional rates
        if countyString == 'Eau Claire': 
            taxRateFloat += EAU_TAX_RATE
        elif countyString == 'Dunn':
            taxRateFloat += DUNN_TAX_RATE
    
    #If we need to do additional calculations for state and/or county
    if(additionTaxOutput):
        taxFloat = taxRateFloat * orderAmountFloat
        totalFloat = totalFloat + taxFloat
        
        #Creates output strings for additional calculations 
        subTotalString = "The subtotal is ${:.2f}".format(orderAmountFloat)
        taxString = "The tax is ${:.2f}".format(taxFloat)
        outputString = "".join([subTotalString,'\n', taxString, '\n'])
    

    #Gets the total Cost and creates an output string
    totalRounded = Decimal(totalFloat).quantize(Decimal('.001'), rounding=ROUND_CEILING)
    totalString = "The total is ${:.2f}".format(totalRounded)
    outputString = ''.join([outputString, totalString])
    print(outputString)    

calcTaxes()
```

# Problem 9 (4 points)

Create a program that compares two strings and determines if the two strings are anagrams. The program should prompt for both input strings and display the output as shown in the example that follows.

## Example Output

<pre>
Enter two strings and I'll tell you if they are anagrams:
Enter the first string: note
Enter the second string: tone
"note" and "tone" are anagrams.
</pre>

## Constraints

- Implement the program using a function called `isAnagram`, which takes in two words as its arguments and returns `true` or `false`. Invoke this function from your main program.
- Check that both words are the same length.

```python
def isAnagram(wordOne, wordTwo):
    #checks to see if two words are the same length
    if(len(wordOne) != len(wordTwo)):
        print("Input words are not the same length!")
        return False;
    else:
        #sorts each word to make the compare easier
        wordOneSort = sortWord(wordOne)
        wordTwoSort = sortWord(wordTwo)
        #if each word sorted is the same then they are anagrams
        if(wordOneSort == wordTwoSort):
            return True
        else:
            return False
        
#This function takes a word and sorts it
def sortWord(word):
    wordList = list(word)
    wordList.sort()
    wordList = ''.join(wordList)
    return wordList

#Main Program
def anagramProg():   
    #Gets input 
    print("Enter two strings and I'll tell you if they are anagrams: ")
    wordOneString = input("Enter the first string: ")
    wordTwoString = input("Enter the second string: ")

    #Builds output section of program
    outputString = '"{0}", and "{1}" '.format(wordOneString,wordTwoString)
    if(isAnagram(wordOneString,wordTwoString)):
        outputString += 'are anagrams.'
    else:
        outputString += 'are not anagrams.'

    print(outputString)
    
anagramProg()
```


# Problem 10 (4 points)

Create a program that determines the complexity of a given password based on these rules:
- A very weak password contains only numbers and is fewer than eight characters.
- A weak password contains only letters and is fewer than eight characters.
- A strong password contains letters and at least one number and is at least eight characters.
- A very strong password contains letters, numbers, and special characters and is at least eight characters.

## Example Output

<pre>
The password '12345' is a very weak password.
The password 'abcdef' is a weak password.
The password 'abc123xyz' is a strong password.
The password '1337h@xor!' is a very strong password.
</pre>

## Constraints

- Create a `passwordValidator` function that takes in the password as its argument and returns a value you can evaluate to determine the password strength. Do not have the function return a string — you may need to support multiple languages in the future.
- Use a single output statement.

```python
#There are a few combinations in this set that didn't fall into any requirements, in those cases it return an invalid combination
def passwordValidator(password): 
    numbers = False
    letters = False
    specials = False
    length = len(password)
    #Set to Negative 1 for all combinations we don't account for like numbers or letters over 8 in length or 
    #Just letters and specials
    returnCode = -1
    #Goes through each letter and figures out what type of character it is
    for a in password:
        if(a.isdigit()):
            numbers = True
        elif(a.isalpha()):
            letters = True
        else:
            specials = True

    #Determines the strength 
    if(numbers and not letters and not specials and length < 8 ):
        returnCode = 1
    elif(letters and not numbers and not specials and length < 8):
        returnCode = 2
    elif(letters and numbers and not specials and length >= 8):
        returnCode = 3
    elif(letters and numbers and specials and length >=8):
        returnCode = 4
    return returnCode

#Main program 
while True:
    #Gets the passwords and sends it to the passwordValidator function 
    password = input("Input a password:")
    passwordStrength = passwordValidator(password)
    
    outputString = "The password '{0}' is a ".format(password)
    
    if passwordStrength == 1:
        outputString += "very weak password."
        break
    elif passwordStrength == 2:
        outputString += "weak password."
        break
    elif passwordStrength == 3:
        outputString += "strong password."
        break
    elif passwordStrength == 4:
        outputString += "very strong password."
        break
    else:
        outputString += "invalid combination. Please try again"
        print(outputString)
    
print(outputString)
```

### Problem 11 (6 points)

Write a program that will help you determine how many months it will take to pay off a credit card balance. The program should ask the user to enter the balance of a credit card and the APR of the card. The program should then return the number of months needed.

The formula for this is

$n = -\frac{1}{30} \times \frac{\log(1 + \frac{b}{p}(1 - (1 + i)^{30}))}{\log(1+i)}$

where

- `n` is the number of months.
- `i` is the daily rate (APR divided by 365).
- `b` is the balance.
- `p` is the monthly payment.

## Example Output

<pre>
What is your balance? 5000
What is the APR on the card (as a percent)? 12
What is the monthly payment you can make? 100

It will take you 70 months to pay off this card.
</pre>

## Constraints

- Prompt for the card’s APR. Do the division internally.
- Prompt for the APR as a percentage, not a decimal.
- Use a function called `calculateMonthsUntilPaidOff`, which accepts the balance, the APR, and the monthly payment as its 
arguments and returns the number of months. Don’t access any of these values outside the function.
- Round fractions of a cent up to the next cent.

```python
#Imports
from math import log, ceil
from decimal import Decimal, ROUND_CEILING

def calculateMonthsUntilPaidOff(balance, apr, monthlyPayment):
    #Get the daily rate
    i = (1 + (apr/100/365))**30 -1
    
    #get the number of months left 
    numerator = -1 *  log(1 - (balance/monthlyPayment) * i)
    denominator = log(1+i)
    monthsLeft = numerator /denominator
    monthsLeftRounded = ceil(monthsLeft)
    return monthsLeftRounded
#main program
def main():
    #Gets inputs
    while True:
        try:
            loanString = input("Input your loan balance: ")
            loanFloat = float(loanString)
            break;
        except:
            print("Input must be a number")
    while True:
        try:
            aprString = input("What is the APR on the card (as a percent)?  ")
            aprFloat = float(aprString)
            break;
        except:
            print("Input must be a number")

    while True:
        try:
            paymentString = input("What is the monthly payment you can make? ")
            paymentFloat = float(paymentString)
            break;
        except:
            print("Input must be a number")        
    #gets number of months left to pay off balance
    numMonths = calculateMonthsUntilPaidOff(loanFloat, aprFloat, paymentFloat)
    
    #output
    outputString = 'It will take you {0} months to pay off this card.'.format(numMonths)
    print(outputString)
main()
```

# Problem 13 (3 points)

Write a program that prompts the user for five numbers and computes the total of the numbers.

## Example Output

<pre>
Enter a number: 1
Enter a number: 2
Enter a number: 3
Enter a number: 4
Enter a number: 5
The total is 15.
</pre>

## Constraints

- The prompting must use repetition, such as a counted loop, not five separate prompts.

```python
def loopCounter():
    #sets counter variables
    counter = 0
    sumTotal = 0.0
    #used to make sure we get 5 inputs
    while True:
        #makes sure we don't get a character somewhere a long the way and mess up our count
        try:
            inputString = input("Enter a number: ")
            inputFloat = float(inputString)
            sumTotal += inputFloat
            counter += 1
            #reached out 5 inputs
            if(counter == 5):
                break
        #If there is an issue with the input the counter isn't incremented so the user must input 5 numbers
        except:
            print("Input a valid Number!!!!")

    #output!
    outputString = "The total is {0}".format(sumTotal)
    print(outputString)
    
loopCounter()
```

# Problem 14 (5 points)

Write a quick calculator that prompts for the rate of return on an investment and calculates how many years it will take to double your investment.

The formula is $years = 72 / r$, where $r$ is the stated rate of return.

## Example Output

<pre>

Sorry. That's not a valid input.
What is the rate of return? ABC
Sorry. That's not a valid input.

- Don’t allow the user to enter 0.
- Don’t allow non-numeric values.

```python
def doubleInvestment():
    while True:
        #makes sure we don't get a character somewhere a long the way and mess up our count
        try:
            inputString = input("What is the rate of return? ")
            inputFloat = float(inputString)
            if inputFloat <= 0:
                print("Sorry. That's not a valid input.")
            else:
                years = 72 / inputFloat
                break
        except:
            print("Sorry. That's not a valid input.")

    #output!
    outputString = "It will take {0} years to double you initial investment".format(years)
    print(outputString)

doubleInvestment()
```

# Problem 15 (3 points)

Create a program that generates multiplication tables for the numbers 0 through 12.

## Example Output

<pre>
       1    2  ...   12
  1    1    2        12
  2    2    4        24
  :
 12   12   24       144
</pre>

## Constraint

- Use a nested loop to complete this program.

```python
def multiplicationTable():
    for row in range(0,13): 
        for col in range(0,13):
            #Prints space in top left corner
            if (row == 0 and col == 0):
                print(' ', end = '\t')
            #Prints first row between 1 - 12
            elif (row == 0):
                print(col, end="\t")
            #Prints first column 1 -12
            elif(col == 0):
                print(row, end='\t')
            #Prints the table
            else:
                print(row*col, end="\t")  
        print('\n')
    
multiplicationTable()
```

# Problem 21 (6 points)

Write a program that prompts for response times from a website in milliseconds. It should keep asking for values until the user enters “done.”

The program should print the average time (mean), the minimum time, the maximum time, and the standard deviation.

To compute the average (mean)
- Compute the sum of all values.
- Divide the sum by the number of values.

To compute the standard deviation
- Calculate the difference from the mean for each number and square it.
- Compute the mean of the squared values.
- Take the square root of the mean.

## Example Output

<pre>
Enter a number: 100
Enter a number: 200
Enter a number: 1000
Enter a number: 300
Enter a number: done

Numbers: 100, 200, 1000, 300
The average is 400.
The minimum is 100.
The maximum is 1000.
The standard deviation is 400.25.
</pre>

## Constraints

- Use loops and sequences to perform the input and mathematical operations.
- Be sure to exclude the “done” entry from the sequence of inputs.
- Be sure to properly convert numeric values to strings.
- Keep the input separate from the processing and the output.

```python
from math import sqrt

def numberStats():
    #Empty list of the input numbers 
    numberList = []
    #Gets all of the numbers
    while True:
        try:
            numberString = input("Enter a number: ")
            numberFloat = float(numberString)
            numberList.append(numberFloat)
        except:
            if numberString == 'done':
                break
            else:
                print("Input a valid number or 'done'")
    #Gets the mean of the numbers
    meanValue = mean(numberList)
    #Gets the std dev of the numbers
    standardDevValue = standardDev(meanValue, numberList)
    
    #Prints the output
    print("Numbers: {0}".format(", ".join(str(x) for x in numberList)))
    print("The average is {0}.".format(meanValue))
    print("The minimum is {0}.".format(min(numberList)))
    print("The maximum is {0}.".format(max(numberList)))
    print("The standard deviation is {0:.2f}.".format(standardDevValue))

#Calculates mean 
def mean(numList):
    totalNums = len(numList)
    runningTotal = 0
    for num in numList:
        runningTotal += num
    meanValue = runningTotal / totalNums
    return meanValue
#Calcs standand dev
def standardDev(mean, numList):
    runningTotal = 0
    variance = 0
    for num in numList:
        difference = num - mean
        differenceSquared = difference ** 2
        runningTotal += differenceSquared
    variance = runningTotal / (len(numList) - 1)
    stdDev = sqrt(variance)
    return stdDev
numberStats()
```

# Problem 30 (3 points)

Create a program that reads in a file and counts the frequency of words in the file. Then construct a histogram displaying the words and the frequency, and display the histogram to the screen.

## Example Output

Given the text file words.txt with this content

> badger badger badger badger mushroom mushroom snake badger badger badger

the program would produce the following output:

<pre>
badger:   *******
mushroom: **
snake:    *
</pre>

## Constraint

- Ensure that the most used word is at the top of the report and the least used words are at the bottom.

```python
from collections import Counter
def wordCount():
    #Opens the words file
    with open('words.txt', 'r') as file:
        #Reads in the entire file
        line = file.read()
        #splits the string
        lineSplit = line.strip().split(' ')
    #Creates a diction writh a word and a count
    c = Counter(lineSplit)
    #Gets the max Length of the elements and adds one to it for printing
    maxLength = max(map(len, c.elements())) + 1
    #prints everything in the list order by the highest number 
    for i in c.most_common(): 
        print("{word:{lea}}: {stars}".format(word = i[0], lea = maxLength , stars = i[1] * '*'))
wordCount()
```
